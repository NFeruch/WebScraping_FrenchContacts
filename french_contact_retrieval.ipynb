```python
from requests_html import AsyncHTMLSession
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import re
import datetime as dt
import os
from IPython.display import clear_output, display
import time
from webdriver_manager.chrome import ChromeDriverManager
```

```python
import http.client
http.client._MAXHEADERS = 1000

from requests_html import AsyncHTMLSession
asession = AsyncHTMLSession()
```

```python
with open('data/notre_equipe.txt', 'r') as f:
    urls = f.read().split('\n')

notre_equipe = pd.DataFrame()

for url in urls:
    URL = 'https://remax1erchoix.com/fr/nos-courtiers.html'
    headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_10_1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/39.0.2171.95 Safari/537.36'}
    response = await asession.get(URL, headers=headers)
    await response.html.arender(timeout=100)
    soup = BeautifulSoup(response.html.html, 'lxml')
    
    data = pd.DataFrame()

    for person in soup.find_all('div', {'class': 'details text-left'}):
        info = {}

        names = [name.replace('\n', '') for name in person.find('h3').text.split(' ') if name != '' and '.' not in name]

        info['First Name'] = names[0].strip().replace(',', '')
        info['Last Name'] = ' '.join(names[1:]).strip().replace(',', '').title()
        try:
            number = person.find('span', {'class': 'far fa-mobile'}).find_next_sibling().text
            info['Cell Number'] = re.sub(r'^(\d{3})', r'(\1)', number)
        except:
            info['Cell Number'] = np.nan
        info['Email'] = person.find('span', {'class': 'far fa-envelope'}).find_next_sibling().text
        info['Website'] = URL
        
        person_info = pd.DataFrame.from_dict(info, orient='index').transpose()
        data = data.append(person_info)
    
    notre_equipe = notre_equipe.append(data).reset_index(drop=True)
    clear_output(wait=True)
    display(notre_equipe)

notre_equipe.to_csv('notre_equipe.csv', index=False, encoding='latin1')
notre_equipe
```
